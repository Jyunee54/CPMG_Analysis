```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 로드
data = np.load('../data/exp_data/exp_data_32.npy')
data_deno = np.load('../data/exp_data/exp_data_32_deno.npy')

# x축은 그냥 포인트 인덱스라고 가정 (τ time에 비례)
plt.figure(figsize=(16, 5))  # 가로 16인치, 세로 5인치
plt.plot(data)
plt.xlabel('Time step')
plt.ylabel('Coherence (Px)')
plt.title('Raw Noised CPMG Signal')
plt.show()

# x축은 그냥 포인트 인덱스라고 가정 (τ time에 비례)
plt.figure(figsize=(16, 5))  # 가로 16인치, 세로 5인치
plt.plot(data_deno)
plt.xlabel('Time step')
plt.ylabel('Coherence (Px)')
plt.title('Raw Denoised CPMG Signal')
plt.show()
```
