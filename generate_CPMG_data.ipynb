```python
import numpy as np
from imports.models import *
from imports.utils import *
import time
import matplotlib.pyplot as plt
```

```python
N_PULSE_32 = 32
N_PULSE_64 = 64
N_PULSE_96 = 96
N_PULSE_128 = 128
N_PULSE_256 = 256
MAGNETIC_FIELD = 403.553            # Unit: Gauss
GYRO_MAGNETIC_RATIO = 1.0705*1000               # Unit: Herts
WL_VALUE = MAGNETIC_FIELD*GYRO_MAGNETIC_RATIO*2*np.pi

# time_data: a time range of simularated CPMG data. (unit: us, resolution: 4ns)
time_data = np.arange(0, 60, 0.004)
```

```python
total_indices = np.load('./data/total_indices/total_indices_v4_full.npy', allow_pickle=True).item()
AB_lists_dic = np.load('./data/AB_target_dic/AB_target_dic_v4.npy', allow_pickle=True).item()
```

```python
total_spins = np.array([
                        [12000, 32000],
                        [15000, 53000],
                        [18000, 11000],
])
```

## N32

```python
M_values = M_list_return(time_data*1e-6, WL_VALUE, total_spins*2*np.pi, N_PULSE_32)
exp_data_32_deno = (1+M_values) / 2
```

```python
plt.figure(facecolor='w', figsize=(25, 5))
plt.plot(time_data, exp_data_32_deno)
plt.title('Simulated CPMG data')
plt.xlabel('us')
plt.ylabel('Px')

plt.figure(facecolor='w', figsize=(25, 5))
plt.title('Simulated partial CPMG data')
plt.plot(time_data[6000:9500], exp_data_32_deno[6000:9500])
plt.xlabel('us')
plt.ylabel('Px')
```

```python
print(time_data)
print(len(time_data))
```

```python
np.save('./data/exp_data/exp_data_32_deno.npy', exp_data_32_deno)
np.save('./data/time_data/time_data_32.npy', time_data)
```

```python
image_width = 100

fig, axes = plt.subplots(facecolor='w', figsize=(15, 5), nrows=1, ncols=3)
for idx, spin in enumerate(total_spins):
  A_index = return_TPk_from_AB(spin[0], spin[1], WL_VALUE, 1)
  model_index = get_model_index(total_indices, A_index, time_thres_idx=time_data.shape[0], image_width=image_width)
  axes[idx].pcolor(exp_data_32_deno[model_index], vmin=0, vmax=1)
  axes[idx].set_title('(A, B) [kHz]: '+'('+str(spin[0])+', ' + str(spin[1])+')')
```
